# Combining Data from Multiple file through LOOP

1 - import required libraries
2 - import the Glob library
3 - create an empty dataframe and set data on that.
4- use loop to match required pattern in all data.
5 - load the excel file in f into the dataframe df
6- appened the contents of df to the dataframe which we created as an empty.

```python
import pandas as pd
import numpy as np
import glob
```

```python
import pandas as pd
import numpy as np
import glob
all_data = pd.DataFrame()
for f in glob.glob("datasets/data*.xlsx"):
    df = pd.read_excel(f)
    all_data = all_data.append(df,ignore_index=True)
all_data.describe()
```
